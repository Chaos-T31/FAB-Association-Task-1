# INFOSYS BUSINESS AUTOMATED DATA ANALYTICS-

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
FILE_PATH = "../Datasets/Mar 22 - Mar 18.csv"
df = pd.read_csv(FILE_PATH, encoding='cp1252')
df.head()
```

```python
df_flip = df.T
display(df_flip)
# df_flip.columns
# df_flip.reset_index(inplace = True)
# df.drop(, axis = 1, inplace = True)
# pd.concat([pd.Series(np.arange(df.shape[0])), df_flip])
df_flip["EQUITIES AND LIABILITIES SHAREHOLDER’S FUNDS"] = df_flip.index
df_flip.index = np.arange(df.shape[1])
df_flip.columns = df_flip.iloc[0]
df_flip = df_flip.iloc[1:]
col1 = df_flip.pop("EQUITIES AND LIABILITIES SHAREHOLDER’S FUNDS")
# df_final = df_flip.iloc[:, :]
# df_final.insert(0, "EQUITIES AND LIABILITIES SHAREHOLDER’S FUNDS", col1)
df_flip.insert(0, "Date-Time", col1)
df_flip.reset_index(inplace = True, drop = True)
```

```python
# plt.output_notebook()
# fig = plt.figure(plot_width = 800, plot_height = 800, x_range = (59000, 69000), y_range = (43000, 52000))
# fig.line(np.array(df_flip['Reserves and Surplus']), np.array(df_flip['Total Current Assets']), line_width = 2, color = "green")
# plt.show(fig)
# df_flip['Reserves and Surplus'], df_flip['Total Current Assets']


df_flip = df_flip.dropna(axis = 1)
# try:
# display(df_flip)
df_final = df_flip.iloc[:, 1:].applymap(lambda x: float(x.split()[0].replace(',', '')))
# display(df_final)
# except: pass
df_final = df_final.astype(float)
# df_final = df_final.apply(pd.to_numeric)
# display(df_final)
# display(df_flip['Date-Time'])
df_final = pd.concat([pd.DataFrame(df_flip['Date-Time']), df_final], axis = 1)
# display(df_final)

# print(df_final.isna().sum())
print(df_final.dtypes)
# print(df_final['Total Current Assets'])
# print(type(df_final['Reserves and Surplus'][0]))
sns.pairplot(data = df_final.iloc[:, 1:], vars = ['Reserves and Surplus', 'Total Shareholders Funds'], hue = 'Total Current Assets', dropna = True)
df_final
```

## Feature-Label Selection-

```python
X = df_final.drop('FOB Value Of Goods', axis = 1)
Y = df_final['FOB Value Of Goods']
display(X)
display(Y)
```

```python
Y.value_counts()
```

## AutoML-

```python
import evalml
from evalml.preprocessing import split_data
```

```python
from evalml.problem_types import ProblemTypes

prob_types = ProblemTypes.all_problem_types
```

```python
X_train, X_test, Y_train, Y_test = split_data(X, Y, problem_type = prob_types[2])
```

```python
X_train.head()
```

```python
from evalml import AutoMLSearch
```

```python
automl = AutoMLSearch(X_train = X_train, y_train = Y_train, problem_type = prob_types[2], max_batches = 1, optimize_thresholds = True)
```

```python
automl.search()
```

```python
automl.rankings
```

```python
best_pipeline = automl.best_pipeline
best_pipeline
```

```python
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python
scores = best_pipeline.score(X_test, Y_test,  objectives = evalml.objectives.get_core_objectives('regression'))
print(scores)
print(f"Accuracy: {scores['Root Mean Squared Error']}")
```

```python
evalml.objectives.get_all_objective_names()
```

```python
automl_auc = AutoMLSearch(X_train = X_train, y_train = Y_train,
                          problem_type = 'regression',
                          objective = 'Root Mean Squared Error',
                          additional_objectives = ['MAE'],
                          max_batches = 1,
                          optimize_thresholds = True)

automl_auc.search()
```

```python
automl_auc.rankings
```

```python
automl_auc.describe_pipeline(automl_auc.rankings.iloc[0]["id"])
```

```python
best_pipeline_auc = automl_auc.best_pipeline
best_pipeline_auc
```

```python
best_pipeline.save('Model.pkl')
```

```python
check_model = automl.load('Model.pkl')
```
