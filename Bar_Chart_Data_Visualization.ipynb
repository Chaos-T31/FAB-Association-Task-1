```python
n = df.shape[0]
CURR_DIR = os.getcwd()
DIR_NAME = "Bar Graphs"
FOLDER_PATH = os.path.join(CURR_DIR, DIR_NAME).replace("\\\\", "\\")
try:
#     os.rmdir(FOLDER_PATH)
    shutil.rmtree(FOLDER_PATH)
except:
    pass
finally:
    os.mkdir(DIR_NAME)
```
